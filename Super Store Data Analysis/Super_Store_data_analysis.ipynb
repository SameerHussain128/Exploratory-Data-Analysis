# VISUALIZATION

```python
# 1. Bar chart of sales by category
import matplotlib.pyplot as plt

# Group sales by category and sum
category_sales = store.groupby('Category')['Sales'].sum()

# Create a bar chart
category_sales.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# 2. Histogram of profit

# Create a histogram of profit
plt.hist(store['Profit'], bins=20)
plt.title('Distribution of Profit')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3. Scatter plot of sales vs discount
# Create a scatter plot
plt.scatter(store['Discount'], store['Sales'])
plt.title('Sales vs. Discount')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.show()
```

```python
# 4. Pie chart of sales by region
# Group sales by region and sum
region_sales = store.groupby('Region')['Sales'].sum()

# Create a pie chart
region_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Distribution by Region')
plt.ylabel('')  # Remove the default ylabel
plt.show()
```
